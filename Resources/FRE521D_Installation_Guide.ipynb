<h2 style="color: #002145;">4. PostgreSQL</h2>

PostgreSQL is our primary relational database for learning SQL.

### 4.1 Download and Install

#### Windows

1. Go to: **[https://www.postgresql.org/download/windows/](https://www.postgresql.org/download/windows/)**
2. Download the installer from EDB.
3. Run the installer and select:
   - PostgreSQL Server
   - pgAdmin 4
   - Command Line Tools
4. Set a password for the postgres superuser. **Remember this password!**
5. Keep the default port (5432).
6. Complete the installation.

#### macOS

Install using Homebrew (recommended):
```bash
brew install postgresql@16
brew services start postgresql@16
```

Or download from: **[https://www.postgresql.org/download/macosx/](https://www.postgresql.org/download/macosx/)**

#### Linux (Ubuntu/Debian)

```bash
sudo apt update
sudo apt install postgresql postgresql-contrib
sudo systemctl start postgresql
```

### 4.2 Verify Installation

Open a terminal and connect to PostgreSQL:
```bash
psql -U postgres
```

Enter your password when prompted. You should see: `postgres=#`

Type `\q` to exit.

### 4.3 Create Course Database

```bash
psql -U postgres
CREATE DATABASE fre521d;
\q
```

<h2 style="color: #002145;">10. Troubleshooting Common Issues</h2>

### Anaconda / Python

**"conda is not recognized" (Windows)**
- Reinstall Anaconda and check "Add to PATH"
- Or manually add Anaconda to PATH in System Environment Variables

**Package installation fails**
- Try: `conda update conda`
- Use pip as fallback: `pip install packagename`

### PostgreSQL

**Cannot connect: "Connection refused"**
- Ensure PostgreSQL service is running
- Windows: Check Services app for "postgresql" service
- macOS: `brew services start postgresql@16`
- Linux: `sudo systemctl start postgresql`

**Password authentication failed**
- Verify you're using the correct postgres password
- Try resetting: `ALTER USER postgres PASSWORD 'newpassword';`

### VS Code

**Python interpreter not found**
- Open Command Palette > Python: Select Interpreter
- If fre521d not listed, restart VS Code
- Verify conda environment exists: `conda env list`

**SQLTools not connecting**
- Ensure PostgreSQL driver extension is installed
- Check connection settings match PostgreSQL configuration

### Google Cloud / BigQuery

**Authentication errors**
- Re-run: `gcloud auth login`
- Re-run: `gcloud auth application-default login`
- Verify project: `gcloud config get-value project`

**BigQuery API errors**
- Ensure BigQuery API is enabled in Cloud Console
- Check billing is set up (even for free tier)

### Git

**"git is not recognized" (Windows)**
- Restart terminal after Git installation
- Verify Git was added to PATH during installation
